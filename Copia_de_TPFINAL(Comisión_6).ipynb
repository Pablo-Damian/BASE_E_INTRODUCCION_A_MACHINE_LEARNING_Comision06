**Diccionario de Datos (*Variables*)**

- **Tournament (*Torneo*):** Esta variable indica el nombre del torneo en el que se jugó el partido. Puede incluir torneos de diferentes niveles y categorías.

- **Matchday (*Día de Partido*):** Representa el número de día en el que se jugó el partido. Es un valor numérico entero que identifica el día del encuentro.

- **Date (*Fecha*):** La variable "Date" representa la fecha en la que tuvo lugar el partido. Está en formato de fecha y puede ser utilizada para analizar el rendimiento de Messi en diferentes momentos del año.

- **Venue (*Lugar*):** Indica si el partido se jugó en el estadio de Messi (casa/local) o en un estadio adversario (fuera/visitante).

- **Club:** Muestra el nombre del club al que pertenece/cía Lionel Messi en el momento del partido.

- **Opponent (*Oponente*):** Representa el nombre del equipo oponente con el que se enfrentó Messi en ese partido.

- **Result (*Resultado*):** Esta variable describe el resultado final del partido en 'formato de resultado' (por ejemplo, "1:2" para indicar un marcador de 1-2).

- **Position (*Posición*):** Indica la posición en la que jugó Messi durante el partido (delantero, centrocampista, etc.).

- **Minute (*Minuto*):** La variable "Minute" muestra el minuto en el que ocurrió un evento relevante durante el partido (por ejemplo, el minuto en que Messi anotó un gol).

- **When Scored (*Momento del Gol*):** Indica el marcador en el momento exacto en que Messi anotó un gol.

- **Goal Type (*Tipo de Gol*):** Describe el tipo de gol que Messi anotó en el partido (por ejemplo, "*Left-footed shot*" para un gol con el pie izquierdo).

- **Assist (*Asistencia*):** Indica quién asistió en el gol anotado por Messi, si aplica.

```python
# Detección de 'Valores Ausentes'
valoresAusentes = dataMessi.isnull().sum()
print(valoresAusentes)
```

```python
# Desactivar las 'advertencias'
import warnings
warnings.filterwarnings("ignore")
# Volver a activar las 'advertencias'
#warnings.filterwarnings("default")

# Detección de Valores Atípicos
# Calcula el rango intercuartil (IQR)
Q1 = dataMessi.quantile(0.25)
Q3 = dataMessi.quantile(0.75)
IQR = Q3 - Q1

# Define límites para los valores atípicos
limiteInferior = Q1 - 1.5 * IQR
limiteSuperior = Q3 + 1.5 * IQR

# Detecta los valores atípicos
valoresAtipicos = ((dataMessi < limiteInferior) | (dataMessi > limiteSuperior)).sum()
print(valoresAtipicos)
```

```python
## Los gráficos del análisis fueron realizados sólo con 5 variables de las 12 disponibles (sólo a modo de ejemplo)
# Configuración del estilo de los gráficos
sns.set(style="whitegrid")

# Gráfico para la variable - 0 - Tournament (Torneo)
plt.figure(figsize=(10, 6))
sns.countplot(x="Tournament", data=dataMessi, palette="Set2")
plt.title("Distribución de Partidos por Torneo")
plt.xticks(rotation=45)
plt.show()

# Gráfico para la variable - 6 - Result (Resultado)
plt.figure(figsize=(20, 6))
sns.countplot(x="Result", data=dataMessi, palette="Set1")
plt.title("Distribución de Resultados de Partidos")
plt.show()

# Gráfico para la variable - 7 - Position (Posición)
plt.figure(figsize=(8, 6))
sns.countplot(x="Position", data=dataMessi, palette="Set3")
plt.title("Distribución de Posiciones Jugadas por Messi")
plt.xticks(rotation=45)
plt.show()

# Histograma de la variable - 8 - Minute (Minuto)
plt.figure(figsize=(40, 6))
sns.histplot(dataMessi["Minute"], bins=20, kde=True, color="purple")
plt.title("Distribución de Minutos en los que Ocurrieron Eventos")
plt.xlabel("Minuto")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico para la variable - 10 - Goal Type (Tipo de Gol)
plt.figure(figsize=(12, 6))
sns.countplot(x="Goal Type", data=dataMessi, palette="Set2")
plt.title("Distribución de Tipos de Goles Anotados por Messi")
plt.xticks(rotation=45)
plt.show()
```

Realizar análisis de matriz correlación y explicar que variable están correlacionadas.


---



```python
### Limitación de la matriz de correlación en este 'conjunto de datos'
##  La 'matriz de correlación' NO puede ser completamente aplicable en este conjunto de datos debido a las siguientes razones:
#   1. La mayoría de las variables son 'categóricas', y la 'matriz de correlación' es más adecuada para 'variables numéricas'.
#   2. Las columnas 'categóricas', como 'Tournament' y 'Goal Type', no son directamente comparables en una 'matriz de correlación'.
##  En lugar de la 'matriz de correlación' se considera un 'gráfico de dispersión' entre dos columnas
columna1 = "Matchday"
columna2 = "Goal Type"
plt.figure(figsize=(18, 6))
sns.scatterplot(data=dataMessi, x=columna1, y=columna2)
plt.title(f"Dispersión entre {columna1} y {columna2}")
plt.xticks(rotation=45)  # Ajusta el ángulo de rotación de las etiquetas en el eje x
plt.xlabel(columna1)
plt.ylabel(columna2)
plt.show()
```

**Explicación del 'Gráfico de Dispersión':**

Cada punto en el gráfico representa una observación en el conjunto de datos. En este caso, la relación entre las variables "*Matchday*" y "*Goal Type*" para el jugador de fútbol Messi.

- **En el eje x (horizontal)**, tenemos la variable **"Matchday"**, que representa el día del partido en el que se anotó un gol.
- **En el eje y (vertical)**, tenemos la variable **"Goal Type"**, que representa el tipo de gol anotado (como 'tiro con el pie izquierdo', 'tiro con el pie derecho', 'cabezazo', 'tiro libre' o 'penal').

- Cada **punto en el gráfico** representa un gol en un partido específico. La posición de cada punto en el gráfico muestra en **qué día del partido se anotó el gol en el eje x**, y **qué tipo de gol fue en el eje y**.

- La **dispersión de los puntos** en el gráfico nos da una idea de cómo se distribuyen los diferentes tipos de goles en relación con los días del partido. Si los puntos se agrupan alrededor de ciertos valores en ambos ejes, podríamos inferir una relación entre esos valores. Por ejemplo: *si vemos que la mayoría de los goles tipo "Penal" se anotaron en los primeros partidos del día ("Matchday"), podríamos inferir que hay una tendencia a anotar penales al comienzo de los partidos.*
