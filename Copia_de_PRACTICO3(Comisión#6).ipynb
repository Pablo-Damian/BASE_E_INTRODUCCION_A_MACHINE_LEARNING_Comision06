##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el número de lanzamientos que se realizarán
numeroLanzamientos = 10000

# Simulamos el lanzamiento del dado generando 'numeroLanzamientos' de muestras aleatorias (con valores del 1 al 6)
muestrasDado = np.random.randint(1, 7, size=numeroLanzamientos)

# Contamos la frecuencia de ocurrencia de cada número en las muestras
frecuenciaNumero2 = np.count_nonzero(muestrasDado == 2)
frecuenciaNumero3 = np.count_nonzero(muestrasDado == 3)
frecuenciaNumero5 = np.count_nonzero(muestrasDado == 5)

# Calculamos la probabilidad aproximada de obtener cada número
probabilidadNumero2 = frecuenciaNumero2 / numeroLanzamientos
probabilidadNumero3 = frecuenciaNumero3 / numeroLanzamientos
probabilidadNumero5 = frecuenciaNumero5 / numeroLanzamientos

print("Probabilidad aproximada de obtener el número 2:", probabilidadNumero2)
print("Probabilidad aproximada de obtener el número 3:", probabilidadNumero3)
print("Probabilidad aproximada de obtener el número 5:", probabilidadNumero5)

# Graficamos el 'histograma' de las probabilidades
nombres = ['2', '3', '5']
probabilidades = [probabilidadNumero2, probabilidadNumero3, probabilidadNumero5]

# Lista de colores en: https://matplotlib.org/stable/_images/sphx_glr_named_colors_003.png
colores = ['blue', 'green', 'red'] # Colores individuales para las barras

plt.bar(nombres, probabilidades, color=colores)
plt.xlabel("Número")
plt.ylabel("Probabilidad")
plt.title("Histograma de Probabilidades del 'Dado Equilibrado'")
plt.grid(axis='y')
plt.show() # Se muestra el gráfico aunque no es necesario en este entorno
```

## ***EJERCICIO 2***

Simular el lanzamiento de un dado NO eqilibrado o cargado de 6 caras, de manera que las chances o probabilidades de salir 1 o 2 sean el doble de obtener otro resultado (3,4,5 o 6)


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Se define el número de lanzamientos que se realizarán
numeroLanzamientos = 10000

# Se Definen las probabilidades de obtener cada número del dado
# Las probabilidades (para 1 y 2) serán el doble que las otras caras
probabilidades = [0.4, 0.4, 0.1, 0.1, 0.1, 0.1]

# Se normalizan las probabilidades para que sumen 1
probabilidades = np.array(probabilidades)
probabilidades /= np.sum(probabilidades)

# Se iimula el lanzamiento del dado generando 'numeroLanzamientos' de muestras aleatorias
muestrasDado = np.random.choice(np.arange(1, 7), size=numeroLanzamientos, p=probabilidades)

# Se cuenta la 'frecuencia de ocurrencia' de cada número en las muestras
frecuenciaNumero2 = np.count_nonzero(muestrasDado == 2)
frecuenciaNumero3 = np.count_nonzero(muestrasDado == 3)
frecuenciaNumero5 = np.count_nonzero(muestrasDado == 5)

# Se calcula la 'probabilidad aproximada' de obtener cada número
probabilidadNumero2 = frecuenciaNumero2 / numeroLanzamientos
probabilidadNumero3 = frecuenciaNumero3 / numeroLanzamientos
probabilidadNumero5 = frecuenciaNumero5 / numeroLanzamientos

print("Probabilidad aproximada de obtener el número 2:", probabilidadNumero2)
print("Probabilidad aproximada de obtener el número 3:", probabilidadNumero3)
print("Probabilidad aproximada de obtener el número 5:", probabilidadNumero5)

# Graficamos el 'histograma de las probabilidades'
nombres = ['2', '3', '5']
probabilidades = [probabilidadNumero2, probabilidadNumero3, probabilidadNumero5]

colores = ['blue', 'green', 'orange'] # Colores individuales para las barras

plt.bar(nombres, probabilidades, color=colores)
plt.xlabel("Número")
plt.ylabel("Probabilidad")
plt.title("Histograma de Probabilidades del 'Dado Cargado'")
plt.grid(axis='y')
plt.show()
```

## EJERCICIO 3

 Realizar la simulación del juego {piedra,papel , tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probailidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidad. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el número de jugadas que se realizarán
NJugadas = 10000

# Se simula el juego "Piedra, Papel o Tijera" generando 'NJugadas' de muestras aleatorias
opciones = ['piedra', 'papel', 'tijera']
eleccionOponente = np.random.choice(opciones, size=NJugadas)

# Contamos la frecuencia de elección de cada opción
frecuenciaPiedra = np.count_nonzero(eleccionOponente == 'piedra')
frecuenciaPapel = np.count_nonzero(eleccionOponente == 'papel')
frecuenciaTijera = np.count_nonzero(eleccionOponente == 'tijera')

# Calculamos la 'probabilidad aproximada' de elección de cada opción
probabilidadPiedra = frecuenciaPiedra / NJugadas
probabilidadPapel = frecuenciaPapel / NJugadas
probabilidadTijera = frecuenciaTijera / NJugadas

print("Probabilidad aproximada de que el oponente elija 'Piedra':", probabilidadPiedra)
print("Probabilidad aproximada de que el oponente elija 'Papel':", probabilidadPapel)
print("Probabilidad aproximada de que el oponente elija 'Tijera':", probabilidadTijera)

# Se grafica el 'histograma de las probabilidades'
nombres = ['piedra', 'papel', 'tijera']
probabilidades = [probabilidadPiedra, probabilidadPapel, probabilidadTijera]

colores = ['blue', 'red', 'green'] # Colores individuales para las barras

plt.bar(nombres, probabilidades, color=colores)
plt.xlabel("Elección")
plt.ylabel("Probabilidad")
plt.title("Probabilidades de 'Elección del Oponente' en el Juego 'Piedra, Papel o Tijera'")
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# EJERCICIO PARTE 1 - PRACTICO Nº2:
ventas_diarias = np.array(
    #Lun,Mar,Mie,Jue,Vie,Sab,Dom
    [[20, 15, 25, 30, 18, 22, 24],  #Producto A
    [12, 20, 14, 8, 15, 18, 16],    #Producto B
    [35, 28, 32, 30, 26, 24, 30],   #Producto C
    [40, 38, 45, 42, 39, 41, 37]    #Producto D
    ]
)

#1 Calcular la media en las ventas por cada producto y la mediana:
mediaPorProducto = np.mean(ventas_diarias, axis=1)
medianaPorProducto = np.median(ventas_diarias, axis=1)

# Etiquetas para el gráfico
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']

# Graficamos la 'media' y la 'mediana' en un 'Gráfico de Barra'
plt.bar(productos, mediaPorProducto, label='Media', color='red')
plt.bar(productos, medianaPorProducto, label='Mediana')
plt.xlabel('Producto')
plt.ylabel('Ventas')
plt.title('Media y Mediana de Ventas por Producto')
plt.legend()
plt.grid(axis='y')
plt.show()

#2 Calcular la media en las ventas por DÍA y la mediana:
mediaPorDia = np.mean(ventas_diarias, axis=0)
medianaPorDia = np.median(ventas_diarias, axis=0)

# Etiquetas para el gráfico
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Graficamos la 'media' y la 'mediana' en un gráfico de 'TRAZO DE LINEAS'
plt.gca().set_facecolor('moccasin')
plt.plot(dias, mediaPorDia, label='Media', marker='o', color='lime')
plt.plot(dias, medianaPorDia, label='Mediana', marker='o', color='red')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title('Media y Mediana de Ventas por Día')
plt.legend()
plt.grid(axis='y')
plt.show()

#3 Realizar una conclusión en función las graficas y sugerir al menos 5 estrategias de venta o inversión de manera de potenciar este negocio:
    #1-Identificar días con ventas bajas y establecer promociones o descuentos para incentivar las ventas en esos días.
    #2-Observar qué productos tienen un bajo rendimiento en ventas y considerar estrategias para mejorar su promoción o lanzar nuevos productos.
    #3-Analizar qué días tienen mayores ventas y considerar aumentar el stock de productos para satisfacer la demanda.
    #4-Ofrecer paquetes o descuentos especiales para aquellos productos que tengan una alta demanda y se vendan bien juntos.
    #5-Implementar estrategias de marketing y publicidad para promover los productos que tengan una alta mediana de ventas en días específicos.

#4 Calcular el desvio estandar de los datos por cada producto:
desvioEstandarPorProducto = np.std(ventas_diarias, axis=1)

# Se obtiene el índice del producto con el mayor y menor desvío estándar
indiceMayorDesvio = np.argmax(desvioEstandarPorProducto)
indiceMenorDesvio = np.argmin(desvioEstandarPorProducto)

# Obtenemos los datos de ventas del producto con mayor y menor desvío estándar
productoMayorDesvio = ventas_diarias[indiceMayorDesvio]
productoMenorDesvio = ventas_diarias[indiceMenorDesvio]

# Etiquetas para el gráfico
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']

# Graficamos los datos de ventas de los productos con mayor y menor desvío estándar
plt.figure(facecolor='skyblue')
plt.plot(dias, productoMayorDesvio, label=f'Producto con mayor desvío', color='orange')
plt.plot(dias, productoMenorDesvio, label=f'Producto con menor desvío', color='darkviolet')
plt.xlabel('Día')
plt.ylabel('Ventas')
plt.title('Ventas de los Productos con Mayor y Menor Desvío Estándar')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la cantidad de encuestados
N100 = 100
N1000 = 1000

# Generamos los DNI para los encuestados
dniJovenes = np.random.randint(10000000, 40000000, size=int(N100*0.7))
dniMayores = np.random.randint(60000000, 90000000, size=int(N100*0.1))
dniEncuestados100 = np.concatenate((dniJovenes, dniMayores))

dniJovenes = np.random.randint(10000000, 40000000, size=int(N1000*0.7))
dniMayores = np.random.randint(60000000, 90000000, size=int(N1000*0.1))
dniEncuestados1000 = np.concatenate((dniJovenes, dniMayores))

# Se generan los datos de los encuestados
ingresosPromedioCor = 258000
ingresosMensuales100 = np.random.normal(loc=ingresosPromedioCor, scale=30000, size=N100)
ingresosMensuales1000 = np.random.normal(loc=ingresosPromedioCor, scale=30000, size=N1000)

gastosMensuales100 = ingresosMensuales100 * (np.random.choice([0.9, 0.8], size=N100, p=[0.8, 0.2]))
gastosMensuales1000 = ingresosMensuales1000 * (np.random.choice([0.9, 0.8], size=N1000, p=[0.8, 0.2]))

cantidadMiembros100 = np.random.normal(loc=5, scale=1, size=N100)
cantidadMiembros1000 = np.random.normal(loc=5, scale=1, size=N1000)

# Calculamos los gastos en servicios de "telefonía" e "internet"
gastosTelefonia100 = gastosMensuales100 * 0.15
gastosTelefonia1000 = gastosMensuales1000 * 0.15

gastosInternet100 = gastosMensuales100 * 0.078
gastosInternet1000 = gastosMensuales1000 * 0.078

# Se calcula el valor medio y desvío estándar de los "gastos mensuales", "ingresos mensuales" y "pagos de servicios"
mediaGastos100 = np.mean(gastosMensuales100)
mediaGastos1000 = np.mean(gastosMensuales1000)

mediaIngresos100 = np.mean(ingresosMensuales100)
mediaIngresos1000 = np.mean(ingresosMensuales1000)

mediaGastosTelefonia100 = np.mean(gastosTelefonia100)
mediaGastosTelefonia1000 = np.mean(gastosTelefonia1000)

mediaGastosInternet100 = np.mean(gastosInternet100)
mediaGastosInternet1000 = np.mean(gastosInternet1000)

desvioGastos100 = np.std(gastosMensuales100)
desvioGastos1000 = np.std(gastosMensuales1000)

desvioIngresos100 = np.std(ingresosMensuales100)
desvioIngresos1000 = np.std(ingresosMensuales1000)

desvioGastosTelefonia100 = np.std(gastosTelefonia100)
desvioGastosTelefonia1000 = np.std(gastosTelefonia1000)

desvioGastosInternet100 = np.std(gastosInternet100)
desvioGastosInternet1000 = np.std(gastosInternet1000)

# Graficamos los datos de los "ingresos" y los "gastos" en gráficos de barra por separado
nombres = ['100 Encuestados', '1000 Encuestados']
ingresosMedios = [mediaIngresos100, mediaIngresos1000]
gastosMedios = [mediaGastos100, mediaGastos1000]

desviosIngresos = [desvioIngresos100, desvioIngresos1000]
desviosGastos = [desvioGastos100, desvioGastos1000]

plt.figure(facecolor='aquamarine')
plt.bar(nombres, ingresosMedios, yerr=desviosIngresos, capsize=10, color='crimson')
plt.xlabel('Cantidad de Encuestados')
plt.ylabel('Ingresos Mensuales')
plt.title('Media y Desvío Estándar de Ingresos Mensuales')
plt.show()

plt.figure(facecolor='aquamarine')
plt.bar(nombres, gastosMedios, yerr=desviosGastos, capsize=10, color='royalblue')
plt.xlabel('Cantidad de Encuestados')
plt.ylabel('Gastos Mensuales')
plt.title('Media y Desvío Estándar de Gastos Mensuales')
plt.show()

# Se calcula la proporción de jóvenes menores a 21 encuestados y mostramos sus datos
cantidadJovenes100 = np.sum(cantidadMiembros100 <= 21)
cantidadJovenes1000 = np.sum(cantidadMiembros1000 <= 21)

proporcionJovenes100 = cantidadJovenes100 / N100
proporcionJovenes1000 = cantidadJovenes1000 / N1000

gastosInternetJovenes100 = np.sum(gastosInternet100[cantidadMiembros100 <= 21])
gastosInternetJovenes1000 = np.sum(gastosInternet1000[cantidadMiembros1000 <= 21])

gastosTelefoniaJovenes100 = np.sum(gastosTelefonia100[cantidadMiembros100 <= 21])
gastosTelefoniaJovenes1000 = np.sum(gastosTelefonia1000[cantidadMiembros1000 <= 21])

print("Proporción de jóvenes menores a 21 en 100 encuestados:", proporcionJovenes100)
print("Proporción de jóvenes menores a 21 en 1000 encuestados:", proporcionJovenes1000)

print("Gasto total en internet de jóvenes menores a 21 en 100 encuestados:", gastosInternetJovenes100)
print("Gasto total en internet de jóvenes menores a 21 en 1000 encuestados:", gastosInternetJovenes1000)

print("Gasto total en telefonía de jóvenes menores a 21 en 100 encuestados:", gastosTelefoniaJovenes100)
print("Gasto total en telefonía de jóvenes menores a 21 en 1000 encuestados:", gastosTelefoniaJovenes1000)

# Se realiza un 'histograma' con el dato del grupo familiar
plt.hist(cantidadMiembros1000, bins=10, alpha=0.5, label='1000 Encuestados', edgecolor='black')
plt.hist(cantidadMiembros100, bins=10, alpha=0.5, label='100 Encuestados', edgecolor='black', color='deeppink')
plt.xlabel('Cantidad de Miembros en el Hogar')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Cantidad de Miembros en el Hogar')
plt.legend()
plt.grid(axis='y')
plt.show()
```
