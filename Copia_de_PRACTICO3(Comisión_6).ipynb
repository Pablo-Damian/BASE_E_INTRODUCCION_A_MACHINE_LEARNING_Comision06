##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el número de lanzamientos que se realizarán
num_lanzamientos = 10000

# Simulamos el lanzamiento del dado generando 'num_lanzamientos' muestras aleatorias (valores del 1 al 6)
muestras_dado = np.random.randint(1, 7, size=num_lanzamientos)

# Contamos la frecuencia de ocurrencia de cada número en las muestras
frecuencia_numero2 = np.count_nonzero(muestras_dado == 2)
frecuencia_numero3 = np.count_nonzero(muestras_dado == 3)
frecuencia_numero5 = np.count_nonzero(muestras_dado == 5)

# Calculamos la probabilidad aproximada de obtener cada número
probabilidad_numero2 = frecuencia_numero2 / num_lanzamientos
probabilidad_numero3 = frecuencia_numero3 / num_lanzamientos
probabilidad_numero5 = frecuencia_numero5 / num_lanzamientos

print("Probabilidad aproximada de obtener el número 2:", probabilidad_numero2)
print("Probabilidad aproximada de obtener el número 3:", probabilidad_numero3)
print("Probabilidad aproximada de obtener el número 5:", probabilidad_numero5)

# Se grafica el histograma de las probabilidades
nombres = ['2', '3', '5']
probabilidades = [probabilidad_numero2, probabilidad_numero3, probabilidad_numero5]

plt.bar(nombres, probabilidades)
plt.xlabel("Número")
plt.ylabel("Probabilidad")
plt.title("Histograma de Probabilidades del Dado")
plt.grid(axis='y')
plt.show() # Se muestra el gráfico aunque no es necesario en este entorno
```

## ***EJERCICIO 2***

Simular el lanzamiento de un dado NO eqilibrado o cargado de 6 caras, de manera que las chances o probabilidades de salir 1 o 2 sean el doble de obtener otro resultado (3,4,5 o 6)


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import numpy as np
import matplotlib.pyplot as plt

# Se define el número de lanzamientos que se realizarán
num_lanzamientos = 10000

# Se Definen las probabilidades de obtener cada número del dado
# Las probabilidades para 1 y 2 serán el doble de las otras caras
probabilidades = [0.4, 0.4, 0.1, 0.1, 0.1, 0.1]

# Se normalizar las probabilidades para que sumen 1
probabilidades = np.array(probabilidades)
probabilidades /= np.sum(probabilidades)

# Se iimula el lanzamiento del dado generando 'num_lanzamientos' muestras aleatorias
muestras_dado = np.random.choice(np.arange(1, 7), size=num_lanzamientos, p=probabilidades)

# Se cuenta la frecuencia de ocurrencia de cada número en las muestras
frecuencia_numero2 = np.count_nonzero(muestras_dado == 2)
frecuencia_numero3 = np.count_nonzero(muestras_dado == 3)
frecuencia_numero5 = np.count_nonzero(muestras_dado == 5)

# Se calcular la probabilidad aproximada de obtener cada número
probabilidad_numero2 = frecuencia_numero2 / num_lanzamientos
probabilidad_numero3 = frecuencia_numero3 / num_lanzamientos
probabilidad_numero5 = frecuencia_numero5 / num_lanzamientos

print("Probabilidad aproximada de obtener el número 2:", probabilidad_numero2)
print("Probabilidad aproximada de obtener el número 3:", probabilidad_numero3)
print("Probabilidad aproximada de obtener el número 5:", probabilidad_numero5)

# Graficamos el histograma de las probabilidades
nombres = ['2', '3', '5']
probabilidades = [probabilidad_numero2, probabilidad_numero3, probabilidad_numero5]

plt.bar(nombres, probabilidades)
plt.xlabel("Número")
plt.ylabel("Probabilidad")
plt.title("Histograma de Probabilidades del Dado Cargado")
plt.grid(axis='y')
plt.show()
```

## EJERCICIO 3

 Realizar la simulación del juego {piedra,papel , tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probailidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidad. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
import numpy as np
import matplotlib.pyplot as plt

# Se Define el número de jugadas que se realizarán
N = 10000

# Se simula el juego "Piedra, Papel o Tijera" generando 'N' muestras aleatorias
opciones = ['piedra', 'papel', 'tijera']
eleccion_oponente = np.random.choice(opciones, size=N)

# Contamos la frecuencia de elección de cada opción
frecuencia_piedra = np.count_nonzero(eleccion_oponente == 'piedra')
frecuencia_papel = np.count_nonzero(eleccion_oponente == 'papel')
frecuencia_tijera = np.count_nonzero(eleccion_oponente == 'tijera')

# Calculamos la probabilidad aproximada de elección de cada opción
probabilidad_piedra = frecuencia_piedra / N
probabilidad_papel = frecuencia_papel / N
probabilidad_tijera = frecuencia_tijera / N

print("Probabilidad aproximada de que el oponente elija 'Piedra':", probabilidad_piedra)
print("Probabilidad aproximada de que el oponente elija 'Papel':", probabilidad_papel)
print("Probabilidad aproximada de que el oponente elija 'Tijera':", probabilidad_tijera)

# Se grafica el "histograma" de las probabilidades
nombres = ['piedra', 'papel', 'tijera']
probabilidades = [probabilidad_piedra, probabilidad_papel, probabilidad_tijera]

plt.bar(nombres, probabilidades)
plt.xlabel("Elección")
plt.ylabel("Probabilidad")
plt.title("Probabilidades de Elección del Oponente en el Juego 'piedra, papel o tijera'")
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la cantidad de encuestados
N_100 = 100
N_1000 = 1000

# Generamos los DNI para los encuestados
dni_jovenes = np.random.randint(10000000, 40000000, size=int(N_100*0.7))
dni_mayores = np.random.randint(60000000, 90000000, size=int(N_100*0.1))
dni_encuestados_100 = np.concatenate((dni_jovenes, dni_mayores))

dni_jovenes = np.random.randint(10000000, 40000000, size=int(N_1000*0.7))
dni_mayores = np.random.randint(60000000, 90000000, size=int(N_1000*0.1))
dni_encuestados_1000 = np.concatenate((dni_jovenes, dni_mayores))

# Se generan datos de encuestados
ingresos_promedio_cba = 258000
ingresos_mensuales_100 = np.random.normal(loc=ingresos_promedio_cba, scale=30000, size=N_100)
ingresos_mensuales_1000 = np.random.normal(loc=ingresos_promedio_cba, scale=30000, size=N_1000)

gastos_mensuales_100 = ingresos_mensuales_100 * (np.random.choice([0.9, 0.8], size=N_100, p=[0.8, 0.2]))
gastos_mensuales_1000 = ingresos_mensuales_1000 * (np.random.choice([0.9, 0.8], size=N_1000, p=[0.8, 0.2]))

cantidad_miembros_100 = np.random.normal(loc=5, scale=1, size=N_100)
cantidad_miembros_1000 = np.random.normal(loc=5, scale=1, size=N_1000)

# Calculamos los gastos en servicios de "telefonía" e "internet"
gastos_telefonia_100 = gastos_mensuales_100 * 0.15
gastos_telefonia_1000 = gastos_mensuales_1000 * 0.15

gastos_internet_100 = gastos_mensuales_100 * 0.078
gastos_internet_1000 = gastos_mensuales_1000 * 0.078

# Se calcula el valor medio y desvío estándar de los gastos mensuales, ingresos mensuales y pagos de servicios
media_gastos_100 = np.mean(gastos_mensuales_100)
media_gastos_1000 = np.mean(gastos_mensuales_1000)

media_ingresos_100 = np.mean(ingresos_mensuales_100)
media_ingresos_1000 = np.mean(ingresos_mensuales_1000)

media_gastos_telefonia_100 = np.mean(gastos_telefonia_100)
media_gastos_telefonia_1000 = np.mean(gastos_telefonia_1000)

media_gastos_internet_100 = np.mean(gastos_internet_100)
media_gastos_internet_1000 = np.mean(gastos_internet_1000)

desvio_gastos_100 = np.std(gastos_mensuales_100)
desvio_gastos_1000 = np.std(gastos_mensuales_1000)

desvio_ingresos_100 = np.std(ingresos_mensuales_100)
desvio_ingresos_1000 = np.std(ingresos_mensuales_1000)

desvio_gastos_telefonia_100 = np.std(gastos_telefonia_100)
desvio_gastos_telefonia_1000 = np.std(gastos_telefonia_1000)

desvio_gastos_internet_100 = np.std(gastos_internet_100)
desvio_gastos_internet_1000 = np.std(gastos_internet_1000)

# Graficamos los datos de los ingresos y los gastos en gráficos de barra por separado
nombres = ['100 Encuestados', '1000 Encuestados']
ingresos_medios = [media_ingresos_100, media_ingresos_1000]
gastos_medios = [media_gastos_100, media_gastos_1000]

desvios_ingresos = [desvio_ingresos_100, desvio_ingresos_1000]
desvios_gastos = [desvio_gastos_100, desvio_gastos_1000]

plt.bar(nombres, ingresos_medios, yerr=desvios_ingresos, capsize=10)
plt.xlabel('Cantidad de Encuestados')
plt.ylabel('Ingresos Mensuales')
plt.title('Media y Desvío Estándar de Ingresos Mensuales')
plt.show()

plt.bar(nombres, gastos_medios, yerr=desvios_gastos, capsize=10)
plt.xlabel('Cantidad de Encuestados')
plt.ylabel('Gastos Mensuales')
plt.title('Media y Desvío Estándar de Gastos Mensuales')
plt.show()

# Se calcula la proporción de jóvenes menores a 21 encuestados y mostramos sus datos
cantidad_jovenes_100 = np.sum(cantidad_miembros_100 <= 21)
cantidad_jovenes_1000 = np.sum(cantidad_miembros_1000 <= 21)

proporcion_jovenes_100 = cantidad_jovenes_100 / N_100
proporcion_jovenes_1000 = cantidad_jovenes_1000 / N_1000

gastos_internet_jovenes_100 = np.sum(gastos_internet_100[cantidad_miembros_100 <= 21])
gastos_internet_jovenes_1000 = np.sum(gastos_internet_1000[cantidad_miembros_1000 <= 21])

gastos_telefonia_jovenes_100 = np.sum(gastos_telefonia_100[cantidad_miembros_100 <= 21])
gastos_telefonia_jovenes_1000 = np.sum(gastos_telefonia_1000[cantidad_miembros_1000 <= 21])

print("Proporción de jóvenes menores a 21 en 100 encuestados:", proporcion_jovenes_100)
print("Proporción de jóvenes menores a 21 en 1000 encuestados:", proporcion_jovenes_1000)

print("Gasto total en internet de jóvenes menores a 21 en 100 encuestados:", gastos_internet_jovenes_100)
print("Gasto total en internet de jóvenes menores a 21 en 1000 encuestados:", gastos_internet_jovenes_1000)

print("Gasto total en telefonía de jóvenes menores a 21 en 100 encuestados:", gastos_telefonia_jovenes_100)
print("Gasto total en telefonía de jóvenes menores a 21 en 1000 encuestados:", gastos_telefonia_jovenes_1000)

# Se realiza un 'histograma' con el dato del grupo familiar
plt.hist(cantidad_miembros_100, bins=10, alpha=0.5, label='100 Encuestados')
plt.hist(cantidad_miembros_1000, bins=10, alpha=0.5, label='1000 Encuestados')
plt.xlabel('Cantidad de Miembros en el Hogar')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Cantidad de Miembros en el Hogar')
plt.legend()
plt.grid(axis='y')
plt.show()
```
